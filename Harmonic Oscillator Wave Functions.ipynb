```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.linalg import eigh_tridiagonal
import warnings
```

```python
N=2000
dy=1/N
y=np.linspace(0,1,N+1)
#defining mL2V:
def mL2V(y):
    return 1000*(y-1/2)**2
#tridiagonal matrix formation
d= 1/dy**2 +mL2V(y)[1:-1]
e= -1/(2*dy**2) * np.ones(len(d)-1)
#tridiagonal matrix solution
w,v= eigh_tridiagonal(d,e)
#we have to transpose the eigenvectors column
v_t=v.T
```

```python
#Eigenfunctions: 
plt.title("Wave functions of harmonic oscillator")
plt.xlabel("$x\longrightarrow$")
plt.ylabel("$\psi_0 (x) \longrightarrow$")
plt.grid()
plt.plot(v_t[0],label="$\psi_0$")
plt.plot(v_t[1],label="$\psi_1$")
plt.plot(v_t[2],label="$\psi_2$")
plt.legend()
plt.show()
```

```python
#probability:
plt.title("Probability amplitudes ")
plt.xlabel("$x\longrightarrow$")
plt.ylabel("$|\psi_0 (x)|² \longrightarrow$")
plt.grid()
plt.plot(v_t[0]**2,label="$\psi_0$")
plt.plot(v_t[1]**2,label="$\psi_1$")
plt.plot(v_t[2]**2,label="$\psi_2$")
plt.legend()
plt.show()
```

```python
#Energy eigenvalues:
plt.title("Numarically solved eigen energies")
plt.xlabel("$Eigenstates \longrightarrow$")
plt.ylabel("$Eigen energies \longrightarrow$")
plt.yticks([0*1e44,0.5*1e44,1.0*1e44,1.5*1e44,2.0*1e44,2.5*1e44], ['0', '1/2 ħ', ' ħ', '3/2 ħ', '2 ħ','5/2 ħ'])
plt.bar(np.arange(0,3,1),w[0:3]/((6.626*10e-23)**2),color="orange")
plt.show()
```

```python
#the wave function found from calculation
#as we are interested in teh shape so we are approximately plotting the function and taking the constants to be 1
x=np.linspace(-4,4,100000)
def psi1(x):
    return (1/np.pi**0.25)*np.exp(-x**2 / 2)
#Wave function
plt.title("Mathematically obtained Wave function of ground state ")
plt.xlabel("$x\longrightarrow$")
plt.ylabel("$\psi_0 (x) \longrightarrow$")
plt.grid()
p1=psi1(x)
plt.plot(x,p1,label="$\psi_0$")
plt.axvline(x=1, color='r', linestyle='--')
plt.axvline(x=-1, color='r', linestyle='--')
plt.legend()
plt.show()
warnings.filterwarnings('ignore')
#probability amplitude
plt.title("Probability amplitude of ground state ")
plt.xlabel("$x\longrightarrow$")
plt.ylabel("$|\psi_0 (x)|² \longrightarrow$")
plt.axvline(x=1, color='r', linestyle='--')
plt.axvline(x=-1, color='r', linestyle='--')
plt.grid()

plt.plot(x,p1**2,label="$|\psi_0 (x)|²$",color="orange")
n=0 
def p_classical(x):
    return np.sqrt(2*n + 1)/ (np.pi* np.sqrt((2*n + 1)-x**2))
plt.plot(x,p_classical(x),label="$\psi_{classical}$")

plt.ylim(0,1)
plt.legend()
plt.show()
```
